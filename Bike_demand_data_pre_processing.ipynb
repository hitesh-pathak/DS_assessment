**Import libs and data from google sheet**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

pd.options.display.max_columns = 999
```

```python
sheet_url = "https://docs.google.com/spreadsheets/d/1WieH-IPBXkKPj46XCEsjAgShaRdZ3A9BTN2zz27jSN4/edit#gid=917200446"
csv_url = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')

df = pd.read_csv(csv_url)

df.head()
```

```python
print(df.shape)
df.info()
```

```python
df = df.rename(columns={'weathersit':'weather',
                       'yr':'year',
                       'mnth':'month',
                       'hr':'hour',
                       'hum':'humidity',
                       'cnt':'count',
                       'dteday':'date'})

date = pd.DatetimeIndex(df['date'])

df['date'] = date.date
df['year'] = date.year
df['month'] = date.month_name()
df['weekday'] = date.day_name()

df
```

```python
print(df['season'].unique())
df['weather'].unique()
```

```python
df['season'] = df['season'].map({1:'s1',2:'s2',3:'s3',4:'s4'})
df['weather'] = df['weather'].map({1:'w1',2:'w2',3:'w3'})
df.sample(5)
```

```python
cat_fields = ["season", "year", "month", "weather", "holiday" ,"workingday"]
drop_cols = ['date', 'instant']
df.drop(drop_cols, axis=1, inplace=True)
df.loc[:, cat_fields] = df[cat_fields].astype('category')
df.info()
```

The above shows that there are no null values in data.
